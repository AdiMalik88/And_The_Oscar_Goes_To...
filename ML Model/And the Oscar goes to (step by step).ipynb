```python
# GG_WIN to Oscars

def check_gg_win(row, df_gg):
    gg_row = df_gg[(df_gg['R_YEAR'] == row['R_YEAR']) &
                   (df_gg['CATEGORY'] == row['CATEGORY']) &
                   (df_gg['NAME'] == row['NAME']) &
                   (df_gg['FILM'] == row['FILM'])]
    if not gg_row.empty:
        return gg_row["G_WIN"].values[0]
    else:
        return 0
```

```python
df_oscars['GG_WIN'] = df_oscars.apply(check_gg_win, df_gg=df_gg, axis=1)
```

```python
# SAG_WIN to Oscars

def check_sag_win(row, df_sag):
    sag_row = df_sag[(df_sag['R_YEAR'] == row['R_YEAR']) &
                   (df_sag['CATEGORY'] == row['CATEGORY']) &
                   (df_sag['NAME'] == row['NAME']) &
                   (df_sag['FILM'] == row['FILM'])]
    if not sag_row.empty:
        return sag_row["S_WIN"].values[0]
    else:
        return 0
```

```python
# Dummy encode the 'CATEGORY' column

df_oscars_dummies = pd.get_dummies(df_oscars, columns=['CATEGORY'])

# Convert O_WIN to integer type

df_oscars_dummies['O_WIN'] = df_oscars_dummies['O_WIN'].astype(int)

# Define features and target variable

X = df_oscars_dummies.drop(columns=['O_WIN','R_YEAR', 'NAME', 'FILM'])
y = df_oscars_dummies['O_WIN']
```

```python
# With Over-Sampling approach #
```

```python
from imblearn.over_sampling import RandomOverSampler

# Instantiate RandomOverSampler

oversampler = RandomOverSampler(random_state=42)

# Resample the data

X_over, y_over = oversampler.fit_resample(X, y)

# Combine the resampled data into a DataFrame

df_resampled_over = pd.concat([X_over, y_over], axis=1)

# Check the class distribution after oversampling

print(df_resampled_over['O_WIN'].value_counts())
```

```python
# With Under-Sampling approach #
```

```python
from imblearn.under_sampling import RandomUnderSampler

undersampler = RandomUnderSampler(random_state=42)

# Resample the data

X_under, y_under = undersampler.fit_resample(X, y)

# Combine the resampled data into a DataFrame

df_resampled_under = pd.concat([X_under, y_under], axis=1)

# Check the class distribution after oversampling

print(df_resampled_under['O_WIN'].value_counts())
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

# Assuming df_resampled_under and df_resampled_over are your balanced datasets

# Define features and target variable for under-sampled dataset
X_under = df_resampled_under.drop(columns=['O_WIN'])
y_under = df_resampled_under['O_WIN']

# Define features and target variable for over-sampled dataset
X_over = df_resampled_over.drop(columns=['O_WIN'])
y_over = df_resampled_over['O_WIN']

# Split the under-sampled dataset into training and testing sets
X_train_under, X_test_under, y_train_under, y_test_under = train_test_split(X_under, y_under, test_size=0.2, random_state=42)

# Split the over-sampled dataset into training and testing sets
X_train_over, X_test_over, y_train_over, y_test_over = train_test_split(X_over, y_over, test_size=0.2, random_state=42)

# Initialize DecisionTreeClassifier for under-sampled dataset
clf_under = DecisionTreeClassifier(random_state=42)

# Train the model on under-sampled data
clf_under.fit(X_train_under, y_train_under)

# Predict on the test set for under-sampled data
y_pred_under = clf_under.predict(X_test_under)

# Evaluate the model for under-sampled data
accuracy_under = accuracy_score(y_test_under, y_pred_under)
print("Accuracy for under-sampled data:", accuracy_under)
print("Classification Report for under-sampled data:")
print(classification_report(y_test_under, y_pred_under))

# Initialize DecisionTreeClassifier for over-sampled dataset
clf_over = DecisionTreeClassifier(random_state=42)

# Train the model on over-sampled data
clf_over.fit(X_train_over, y_train_over)

# Predict on the test set for over-sampled data
y_pred_over = clf_over.predict(X_test_over)

# Evaluate the model for over-sampled data
accuracy_over = accuracy_score(y_test_over, y_pred_over)
print("\nAccuracy for over-sampled data:", accuracy_over)
print("Classification Report for over-sampled data:")
print(classification_report(y_test_over, y_pred_over))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, classification_report

# Define the parameter grid to search
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 10, 20, 30, 40, 50],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# List of imbalance datasets
datasets = [(X_train_under, y_train_under, X_test_under, y_test_under, "Under-sampled"),
            (X_train_over, y_train_over, X_test_over, y_test_over, "Over-sampled")]

for X_train, y_train, X_test, y_test, imbalance_type in datasets:
    print(f"Grid Search for {imbalance_type} Data")
    # Initialize grid search
    grid_search = GridSearchCV(estimator=DecisionTreeClassifier(random_state=42),
                               param_grid=param_grid,
                               scoring='accuracy',
                               cv=5,
                               verbose=1)

    # Perform grid search
    grid_search.fit(X_train, y_train)

    # Get the best parameters and the best estimator
    best_params = grid_search.best_params_
    best_estimator = grid_search.best_estimator_

    print("Best Parameters:", best_params)

    # Predict on the test set using the best estimator
    y_pred = best_estimator.predict(X_test)

    # Evaluate the model
    accuracy = accuracy_score(y_test, y_pred)
    print("Accuracy:", accuracy)

    # Print classification report
    print("Classification Report:")
    print(classification_report(y_test, y_pred))
```

```python
# Predict probabilities
y_new_proba = rf_classifier.predict_proba(X_new_encoded)[:, 1]  # Get the probability of being class 1 (win)

# Add probabilities to the dataframe
df_oscar_goes_to_copy['Win_Probability'] = y_new_proba

# Reset Predicted_O_WIN to 0 to ensure clean slate for assigning winners based on probability
df_oscar_goes_to_copy['Predicted_O_WIN'] = 0

# Iterate over each category to assign the win to the highest probability nominee
for category in df_oscar_goes_to_copy['CATEGORY'].unique():
    # Filter dataframe by category
    df_category = df_oscar_goes_to_copy[df_oscar_goes_to_copy['CATEGORY'] == category]
    
    # Find the index of the nominee with the highest win probability in this category
    winner_index = df_category['Win_Probability'].idxmax()
    
    # Assign the win to this nominee
    df_oscar_goes_to_copy.at[winner_index, 'Predicted_O_WIN'] = 1

# Now, df_oscar_goes_to_copy contains a 'Predicted_O_WIN' column where the most likely winners are marked with a 1
```
